```python
import pandas as pd
df = pd.read_csv(r'C:\Users\PRANAM\Desktop\Research\Best2.csv')
df.head()
```

```python
df[df.Class==2].head()
target_names = ['','M','B']
df['attack_type'] = df.Class.apply(lambda x: target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df1 = df[df.Class==1]
df2 = df[df.Class==2]
```

```python
df1.head()
```

```python
plt.xlabel('radius_mean')
plt.ylabel('texture_mean')
plt.scatter(df1['radius_mean'],df1['texture_mean'],color='green',marker='+')
plt.scatter(df2['radius_mean'],df2['texture_mean'],color='blue',marker='.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Class','attack_type'],axis='columns')
X.head()
```

```python
y = df.Class
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
percentage = model.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix 
res = confusion_matrix(y_test,predictions)
print("Confusion Matrix")
print(res)
print(f"Test Set: {len(X_test)}")
print(f"Accuracy = {percentage*100} %")
```
